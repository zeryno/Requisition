```python
import os
from pathlib import Path
import win32com.client as win32

# 文件路径配置
input_excel_path = r"D:\ZHR\Requisition\2025.06\CRJE-2025-06 机关领料单-1.xlsx"  # 输入的 Excel 文件路径
output_folder_name = "1库领料单PDF"  # 输出 PDF 的文件夹名称

# 检查路径是否存在
file_path = Path(input_excel_path)
if not file_path.exists():
    # 提示路径错误并退出
    raise FileNotFoundError(f"输入的 Excel 文件不存在：{file_path}")

# 创建输出文件夹
output_folder = file_path.parent / output_folder_name
if not output_folder.exists():
    output_folder.mkdir(parents=True)

# 打开 Excel 应用
excel_app = win32.Dispatch("Excel.Application")
excel_app.Visible = False  # 不显示 Excel 窗口

try:
    # 打开工作簿
    workbook = excel_app.Workbooks.Open(str(file_path))

    # 遍历所有子表
    for sheet in workbook.Sheets:
        sheet_name = sheet.Name  # 子表名称

        # 生成 PDF 文件名：子表名称.pdf
        pdf_filename = f"{sheet_name}.pdf"
        pdf_output_path = output_folder / pdf_filename

        # 导出子表为 PDF
        sheet.ExportAsFixedFormat(0, str(pdf_output_path))  # 0 表示 PDF 格式
        print(f"成功导出：{pdf_output_path}")

except Exception as e:
    print(f"发生错误：{e}")
finally:
    # 关闭工作簿和 Excel 应用
    workbook.Close(SaveChanges=False)
    excel_app.Quit()

print(f"所有子表 PDF 已生成，结果保存在：{output_folder}")
```
